```python
def frequencyDistribution(value, freq):
    length = len(value)
    max_val = np.max(value)
    min_val = np.min(value)
    delta = max_val - min_val
    freq_range = delta / freq

    print(min_val, max_val, delta, freq_range)

    divisions = {}
    lower_range = min_val
    for j in range(freq):
        divisions[lower_range] = 0
        lower_range +=  freq_range

    for i in range(length):
        lower_range = min_val
        for j in range(freq):
            if(lower_range <= value.iloc[i] and value.iloc[i] < lower_range + freq_range):
                divisions[lower_range] += 1
                break
            lower_range += freq_range
    return divisions


dist = frequencyDistribution(tata_stocks.Low, 10)
```
