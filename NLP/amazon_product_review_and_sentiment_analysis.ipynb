```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv("../../Datasets/Amazon_Product_Review_Dataset/amazon.csv")
```

```python
reviews = data["reviewText"]
labels = data["overall"]
```

```python
vocab_size = 10000
max_length = 100
text_vec_layer = tf.keras.layers.TextVectorization(vocab_size,standardize = "lower",output_sequence_length=max_length)
text_vec_layer.adapt([f"{r}" for r in reviews])
#encoded = text_vec_layer(review)[0]
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder();
enc.fit(labels)
output = pd.get_dummies(data['overall'])
```

```python
X_train = tf.constant([f"{r}" for r in reviews[:4000]])
X_test = tf.constant([f"{r}" for r in reviews[4000:]])
y_train = tf.constant(output[:4000])
y_test = tf.constant(output[4000:])
```

```python

X_test[1:5]
```

```python
embed_size = 128
tf.random.set_seed(42)
model = tf.keras.Sequential([
    text_vec_layer,
    tf.keras.layers.Embedding(vocab_size, embed_size),
    tf.keras.layers.LSTM(128),
    tf.keras.layers.Dense(64, activation= 'relu'),
    tf.keras.layers.Dense(64, activation= 'relu'),
    tf.keras.layers.Dense(5, activation= 'softmax')
])
```

```python
model.compile(loss="binary_crossentropy", optimizer="nadam", metrics=['accuracy'])
```

```python
model.summary()
```

```python

model.fit(X_train,y_train, validation_data=(X_test,y_test),epochs=10,batch_size=24)
```
